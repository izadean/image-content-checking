```python
EXAMPLES = [
    "A bicycle replica with a clock as the front wheel.",
    "A black Honda motorcycle parked in front of a garage."
]
```

# Explore Embeddings and its Shape

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("avsolatorio/GIST-small-Embedding-v0", revision=None)

embeddings = model.encode(EXAMPLES, convert_to_tensor=True)
type(embeddings), embeddings.shape
```

# Simple Speed Test

```python
import time

st = time.time()
model.encode(EXAMPLES*10, convert_to_tensor=True)
end = time.time()

embedding_time = end - st

avg_embedding_time_per_sentence = embedding_time / (len(EXAMPLES)*10)
avg_embedding_time_per_sentence  # On Titan T4 GPU
```
